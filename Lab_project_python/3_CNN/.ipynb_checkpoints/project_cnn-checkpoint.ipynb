# Spliting data

```python
import os, shutil
import numpy as np
```

```python
classes_list = ('airplane','car','cat','dog','flower','fruit','motorbike','person')
```

```python
data_directory = '../DATA/natural_images/'
train_dir = data_directory + 'train/'
test_dir  = data_directory + 'test/'
```

```python
for i_class in classes_list:
    source = data_directory + i_class
    
    train_dir_with_class = data_directory + 'train/' + i_class
    if not os.path.exists(train_dir_with_class):
        os.makedirs(train_dir_with_class)
    
    test_dir_with_class  = data_directory + 'test/' + i_class
    if not os.path.exists(test_dir_with_class):
        os.makedirs(test_dir_with_class)
        
    files = os.listdir(source)
    for i_file in files:
        if np.random.rand(1) < 0.15:
            shutil.copy(source + '/' + i_file, test_dir_with_class + '/' + i_file)
        else:
            shutil.copy(source + '/' + i_file, train_dir_with_class + '/' + i_file)
```

# Data visualisation

```python
import matplotlib.pyplot as plt
import cv2
```

```python
fig = plt.figure(figsize=(24, 12))

for i, i_class in enumerate(classes_list):
    file = os.listdir(train_dir + i_class)[200]
    
    image = cv2.imread(train_dir + i_class + '/' + file)
    print(image.shape)
    
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    fig.add_subplot(2, 4, i+1)
    plt.imshow(image)
# in python script
plt.show()
```

# Prepairing data

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
image_gen = ImageDataGenerator(
    rotation_range=20, # rotate the image 20 degrees
    width_shift_range=0.1, # Shift the pic width by a max of 10%
    height_shift_range=0.1, # Shift the pic height by a max of 10%
    rescale=1/255, # Rescale the image by normalzing it.
    shear_range=0.15, # Shear means cutting away part of the image (max 15%)
    zoom_range=0.15, # Zoom in by 15% max
    horizontal_flip=True, # Allo horizontal flipping
    fill_mode='nearest' # Fill in missing pixels with the nearest filled value
                            )
```

```python
fig = plt.figure(figsize=(24, 12))

for i, i_class in enumerate(classes_list):
    file = os.listdir(train_dir + i_class)[200]
    
    image = cv2.imread(train_dir + i_class + '/' + file)
    print(image.shape)
    
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    fig.add_subplot(2, 4, i+1)
    plt.imshow(image_gen.random_transform(image))
# in python script
plt.show()
```

```python
image_gen.flow_from_directory(train_dir)
```

```python
image_gen.flow_from_directory(test_dir)
```

```python
image_shape = (150,150,3)
```

# Creating model

```python
from keras.models import Sequential
from keras.layers import Activation, Dropout, Flatten, Dense, Conv2D, MaxPooling2D
```

```python
model_1 = Sequential()

model_1.add(Conv2D(filters=32, kernel_size=(3,3),input_shape=(150,150,3), activation='relu',))
model_1.add(MaxPooling2D(pool_size=(2, 2)))

model_1.add(Conv2D(filters=64, kernel_size=(3,3),input_shape=(150,150,3), activation='relu',))
model_1.add(MaxPooling2D(pool_size=(2, 2)))

model_1.add(Conv2D(filters=64, kernel_size=(3,3),input_shape=(150,150,3), activation='relu',))
model_1.add(MaxPooling2D(pool_size=(2, 2)))


model_1.add(Flatten())


model_1.add(Dense(128)) # ,activation=tf.nn.relu
model_1.add(Activation('relu'))

# Dropouts help reduce overfitting by randomly turning neurons off during training.
# Here we say randomly turn off 50% of neurons.
model_1.add(Dropout(0.5))

# Last layer
model_1.add(Dense(8))
model_1.add(Activation('sigmoid'))

# optimizer=tf.train.AdamOptimizer(),
model_1.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model_1.summary()
```

```python
batch_size = 16
```

```python
train_image_gen = image_gen.flow_from_directory(train_dir,
                                               target_size=image_shape[:2],
                                               batch_size=batch_size,
                                               class_mode='categorical')
```

```python
test_image_gen = image_gen.flow_from_directory(test_dir,
                                               target_size=image_shape[:2],
                                               batch_size=batch_size,
                                               class_mode='categorical')
```

```python
train_image_gen.class_indices
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
result_model_1 = model_1.fit_generator(
                                    train_image_gen,epochs=100,
                                    steps_per_epoch=100,
                                    validation_data=test_image_gen,
                                    validation_steps=12)
```

```python
model_1.save('model_1.h5')
```

```python
model_2 = Sequential()

model_2.add(Conv2D(filters=32, kernel_size=(4,4),input_shape=(150,150,3), activation='relu',))
model_2.add(MaxPooling2D(pool_size=(2, 2)))

model_2.add(Conv2D(filters=64, kernel_size=(4,4),input_shape=(150,150,3), activation='relu',))
model_2.add(MaxPooling2D(pool_size=(2, 2)))

model_2.add(Conv2D(filters=128, kernel_size=(4,4),input_shape=(150,150,3), activation='relu',))
model_2.add(MaxPooling2D(pool_size=(2, 2)))


model_2.add(Flatten())

model_2.add(Dense(64))
model_2.add(Activation('relu'))

model_2.add(Dense(128))
model_2.add(Activation('relu'))

# Dropouts help reduce overfitting by randomly turning neurons off during training.
# Here we say randomly turn off 50% of neurons.
model_2.add(Dropout(0.5))

# Last layer
model_2.add(Dense(8))
model_2.add(Activation('softmax'))

#
model_2.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
result_model_2 = model_2.fit_generator(
                                    train_image_gen,epochs=100,
                                    steps_per_epoch=100,
                                    validation_data=test_image_gen,
                                    validation_steps=12)
```

```python
model_2.save('model_2.h5')
```
