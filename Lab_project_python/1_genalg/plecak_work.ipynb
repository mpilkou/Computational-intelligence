```python
from pyeasyga.pyeasyga import GeneticAlgorithm
```

```python
plecakGenAlg = GeneticAlgorithm(
    df,
    population_size=200,generations=100,
    mutation_probability=0.05,elitism=True,maximise_fitness=True)
```

```python
def create_individual(data):
    return [random.randint(0, 1) for _ in range(len(data))]

plecakGenAlg.create_individual = create_individual
```

```python
def crossover(parent_1, parent_2):
    index = random.randrange(1, len(parent_1))
    child_1 = parent_1[:index] + parent_2[index:]
    child_2 = parent_2[:index] + parent_1[index:]
    return child_1, child_2
```

```python
plecakGenAlg.crossover_function = crossover
```

```python
def mutate(individual):
    mutate_index = random.randrange(len(individual))
    if individual[mutate_index] == 0:
        individual[mutate_index] == 1
    else:
        individual[mutate_index] == 0
```

```python
plecakGenAlg.mutate_function = mutate
```

```python
def selection(population):
    return random.choice(population)

plecakGenAlg.selection_function = selection
```

```python
def fitness (individual, data):
    calkowita_wartosc_chr = np.matmul(individual , data["wartosc"])
    calkowita_waga_chr = np.matmul(individual ,data.waga)
    if (calkowita_waga_chr > plecaklimit):
        return 0 
    else:
        print(calkowita_wartosc_chr)
        return calkowita_wartosc_chr
```

```python
plecakGenAlg.fitness_function = fitness
```

```python
plecakGenAlg.run()
```
